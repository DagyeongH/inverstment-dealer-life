### 모델 테스트

```python
import pandas as pd
import re
import numpy as np
import pickle
```

```python
# 단어 사전 불러오기
dict_loc = '/content/drive/MyDrive/자료/2023BigContest/Model/SB_dict.pkl'
with open (dict_loc, 'rb') as f:
  vocab = pickle.load(f)
```

```python
tfidf = pd.read_csv('/content/drive/MyDrive/자료/2023BigContest/Data/SB_tf_idf.csv')
```

```python
df = pd.read_csv('/content/drive/MyDrive/자료/2023BigContest/Model/model_SB_data.csv')
```

```python
df2 = pd.merge(tfidf, df)
```

```python
idx = df2.loc[df2['일자']=='2022-01-27', 'Unnamed: 0']
se = tfidf.iloc[idx.values[0]]
```

```python
# 뉴스 요약
import feedparser
from newspaper import Article
from summa.summarizer import summarize

url = 'https://www.sedaily.com/NewsView/26111LK22J'
article = Article(url, language='ko')
article.download()
article.parse()

NewsFeed = article.text
NewsSum = summarize(NewsFeed)

# print('제목\n',article.title)
# print('뉴스피드\n',NewsFeed)
# print('뉴스서머리\n',NewsSum)
```

```python
model_loc = '/content/drive/MyDrive/자료/2023BigContest/Model/model_SB.pickle'
with open(model_loc,'rb') as fr:
    model_sb = pickle.load(fr)
```

```python
okt = Okt()
newssum_okt = okt.nouns(NewsSum)
```

```python
sum = 0
cnt = 0
for j in range(len(newssum_okt)):
  if newssum_okt[j] in vocab and len(newssum_okt[j]) > 1:
    # print(newssum_okt[j], vocab[newssum_okt[j]])
    cnt += 1
    if se[newssum_okt[j]] != 0:
      sum += (vocab[newssum_okt[j]] * se[newssum_okt[j]])
    else:
      sum += vocab[newssum_okt[j]]

score = sum/cnt

model_sb.predict([[score]])
```
