```python
import pandas as pd
import os
import glob
```

```python
save_loc = '/content/drive/MyDrive/자료/2023BigContest/Data'
```

### 뉴스 파일 concat

```python
dirs = glob.glob(r'/content/drive/MyDrive/자료/2023BigContest/Data/SB*.xlsx')
```

```python
for idx, dir in enumerate(dirs):
  print(dir)
  n_df = pd.read_excel(dir)
  if idx == 0:
    df = n_df
    continue
  df = pd.concat([df, n_df], axis=0)

print(len(df))
save_file = os.path.join(save_loc, 'SB_20180101-20221231.csv')
df.to_csv(save_file)
```
